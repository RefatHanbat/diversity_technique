```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
```

```python
ip = np.random.randn(N) > 0.5 
```

```python
s = 2 * ip - 1 
s = np.array(s)
```

```python
eb_no_db = np.arange(0,30)
```

```python
bit_error_rate = np.zeros(len(eb_no_db))
```

```python
def ZF_DFB():
    for jj in range (len(eb_no_db)):
        
        nErr = 0 
        
        for ii in range (N // 2):
            
            h11 = (1 / np.sqrt(2)) * (np.random.rand() + 1j * np.random.rand())
            
            h12 = (1 / np.sqrt(2)) * (np.random.rand() + 1j * np.random.rand())
            
            h21 = (1 / np.sqrt(2)) * (np.random.rand() + 1j * np.random.rand())
            
            h22 = (1 / np.sqrt(2)) * (np.random.rand() + 1j * np.random.rand())
            
            n11 = (1 / np.sqrt(2)) * (np.random.rand() + 1j * np.random.rand())
            
            n21 = (1 / np.sqrt(2)) * (np.random.rand() + 1j * np.random.rand())
            
            H = np.array([[h11, h12],[h21, h22]])
            
            x1 = s[2 *ii]
            
            x2 = s[2 * ii + 1]
            
            x = np.array([[x1],[x2]])
            
            n = np.array([[n11],[n21]])
            
            y = np.matmul(H,x) + n * 10**(-eb_no_db[jj] / 20 )
            
            power_x1 = np.linalg.norm([h11],ord=2)**2 + np.linalg.norm([h21],ord=2)**2
            
            power_x2 = np.linalg.norm([h12],ord=2)**2 + np.linalg.norm([h22],ord=2)**2
            
            H_conjugate = np.conjugate(H)
            
            H_Hermitian = np.transpose(H_conjugate)
            
            H_Hermitian_H = np.matmul(H_Hermitian,H)
            
            H_inverse = np.linalg.inv(H_Hermitian_H)
            
            W = np.matmul(H_inverse, H_Hermitian)
            
            y_tilde = np.matmul(W, y)
            
            
            
            if power_x1 > power_x2 :
                
                r = y_tilde - np.dot(np.array([[h11],[h21]]),x1)
                
                x2_tilde =(np.dot(H_Hermitian,r) / H_Hermitian_H) 
                
                print(x2_tilde > 0)
                
            else:
                r = y_tilde - np.dot(np.array([[h12],[h22]]),x1)
                
                x1_tilde =(np.dot(H_Hermitian,r) / H_Hermitian_H) 
                
                print(x1_tilde > 0)
                
                
```
