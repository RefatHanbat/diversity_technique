```python
import numpy as np
import matplotlib.pyplot as plt

N = int(1e6)  # number of bits or symbols
Eb_N0_dB = np.arange(0, 26)  # multiple Eb/N0 values

nErr = np.zeros(len(Eb_N0_dB))  # error count for each configuration

for ii in range(len(Eb_N0_dB)):
    # Transmitter
    ip = np.random.rand(N) > 0.5  # generating 0, 1 with equal probability
    s = 2 * ip - 1  # BPSK modulation: 0 -> -1; 1 -> 0

    # Alamouti STBC
    sCode = np.zeros((2, N))
    sCode[:, 0::2] = (1 / np.sqrt(2)) * np.reshape(s, (2, N // 2))  # [x1 x2 ...]
    sCode[:, 1::2] = (1 / np.sqrt(2)) * (np.kron(np.ones((1, N // 2)), np.array([[-1], [1]])) * np.flipud(np.reshape(np.conj(s), (2, N // 2))))  # [-x2* x1* ....]

    h = (1 / np.sqrt(2)) * (np.random.randn(1, N) + 1j * np.random.randn(1, N))  # Rayleigh channel
    hMod = np.kron(np.reshape(h, (2, N // 2)), np.ones((1, 2)))  # repeating the same channel for two symbols

    n = (1 / np.sqrt(2)) * (np.random.randn(1, N) + 1j * np.random.randn(1, N))  # white Gaussian noise, 0 dB variance

    # Channel and noise Noise addition
    y = np.sum(hMod * sCode, axis=0) + 10 ** (-Eb_N0_dB[ii] / 20) * n

    # Receiver
    yMod = np.kron(np.reshape(y, (2, N // 2)), np.ones((1, 2)))  # [y1 y1 ... ; y2 y2 ...]
    yMod[1, :] = np.conj(yMod[1, :])  # [y1 y1 ... ; y2* y2*...]

    # forming the equalization matrix
    hEq = np.zeros((2, N), dtype=np.complex64)
    hEq[:, 0::2] = np.reshape(h, (2, N // 2))  # [h1 0 ... ; h2 0...]
    hEq[:, 1::2] = np.kron(np.ones((1, N // 2), dtype=np.complex64), np.array([[1], [-1]], dtype=np.complex64)) * np.flipud(np.reshape(h, (2, N // 2)))  # [h1 h2 ... ; h2 -h1 ...]
    hEq[0, :] = np.conj(hEq[0, :])  # [h1* h2* ... ; h2 -h1 .... ]
    hEqPower = np.sum(hEq * np.conj(hEq), axis=0)

    yHat = np.sum(hEq * yMod, axis=0) / hEqPower  # [h1*y1 + h2y2*, h2*y1 -h1y2*, ... ]
    yHat[1::2] = np.conj(yHat[1::2])

    # receiver - hard decision decoding
    ipHat = np.real(yHat) > 0

    # counting the errors
    nErr[ii] = np.sum(ip != ipHat)

simBer = nErr / N  # simulated bit error rate
EbN0Lin = 10 ** (Eb_N0_dB / 10)
theoryBer_nRx1 = 0.5 * (1 - 1 * (1 + 1 / EbN0Lin) ** (-0.5))

p = 1 / 2 - 1 / 2 * (1 + 1 / EbN0Lin) ** (-1 / 2)
theoryBerMRC_nRx2 = p ** 2 * (1 + 2 * (1 - p))

pAlamouti = 1 / 2 - 1 / 2 * (1 + 2 / EbN0Lin) ** (-1 / 2)
theoryBerAlamouti_nTx2_nRx1 = pAlamouti ** 2 * (1 + 2 * (1 - pAlamouti))

plt.figure()
plt.semilogy(Eb_N0_dB, theoryBer_nRx1, 'bp-', linewidth=2)
plt.semilogy(Eb_N0_dB, theoryBerMRC_nRx2, 'kd-', linewidth=2)
plt.semilogy(Eb_N0_dB, theoryBerAlamouti_nTx2_nRx1, 'c+-', linewidth=2)
plt.semilogy(Eb_N0_dB, simBer, 'mo-', linewidth=2)
plt.axis([0, 25, 1e-5, 0.5])
plt.grid(True)
plt.legend(['theory (nTx=1,nRx=1)', 'theory (nTx=1,nRx=2, MRC)', 'theory (nTx=2, nRx=1, Alamouti)', 'sim (nTx=2, nRx=1, Alamouti)'])
plt.xlabel('Eb/No, dB')
plt.ylabel('Bit Error Rate')
plt.title('BER for BPSK modulation with Alamouti STBC (Rayleigh channel)')
plt.show()
```
